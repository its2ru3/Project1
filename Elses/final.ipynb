````python
n = qc.width()
g = gate_counts(qc)
h = list(instruction.operation.name for index,
          instruction in enumerate(qc.data)).count('h')
d = qc.depth()
t = n+h
````

````python
def get_statevector(qc, initial_state):
    terms, wire_array = create_poly(qc, n)
    ivs = [j[0] for j in wire_array]
    ovs = [j[-1] for j in wire_array]
    print("Input variables are: ", ivs)
    print("Output variables are: ", ovs)
    ttb = truthtable(terms, n, t, initial_state, ivs, np)
    return statevector_(ttb, n, t, ovs, np)
# counts = {}
````

````python
import pandas as pd

df = pd.read_csv('program_data.csv')
df
````

````python
## When n is CONSTANT

import matplotlib.pyplot as plt
import seaborn as sns
# Set Seaborn style
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Plot CPU times with varying 'h'
sns.lineplot(data=df, x='h', y='cpu_time_poly', marker='o', ax=axes[0], label='cpu_time_poly')
sns.lineplot(data=df, x='h', y='cpu_time_ddsim', marker='x', ax=axes[0], label='cpu_time_ddsim')
sns.lineplot(data=df, x='h', y='cpu_time_aer', marker='^', ax=axes[0], label='cpu_time_aer')
axes[0].set_title('CPU Time vs h')
axes[0].set_xlabel('h')
axes[0].set_ylabel('CPU Time (s)')

# Plot Wall times with varying 'h'
sns.lineplot(data=df, x='h', y='wall_time_poly', marker='o', ax=axes[1], label='wall_time_poly')
sns.lineplot(data=df, x='h', y='wall_time_ddsim', marker='x', ax=axes[1], label='wall_time_ddsim')
sns.lineplot(data=df, x='h', y='wall_time_aer', marker='^', ax=axes[1], label='wall_time_aer')
axes[1].set_title('Wall Time vs h')
axes[1].set_xlabel('h')
axes[1].set_ylabel('Wall Time (s)')

# Improve layout and aesthetics
plt.tight_layout()
plt.show()
````

````python
# Set Seaborn style
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Plot CPU times with varying 'h' and add secondary y-axis for 'n'
ax1 = axes[0]
sns.lineplot(data=df, x='h', y='cpu_time_poly', hue='n', marker='o', ax=ax1, legend=None)
sns.lineplot(data=df, x='h', y='cpu_time_ddsim', hue='n', marker='x', ax=ax1, legend=None)
sns.lineplot(data=df, x='h', y='cpu_time_aer', hue='n', marker='^', ax=ax1, legend=None)
ax1.set_title('CPU Time vs h')
ax1.set_xlabel('h')
ax1.set_ylabel('CPU Time (s)')
ax1_2 = ax1.twinx()  # Create secondary y-axis
sns.lineplot(data=df, x='h', y='n', ax=ax1_2, color='grey', linestyle='--', marker='D')
ax1_2.set_ylabel('n')

# Plot Wall times with varying 'h' and add secondary y-axis for 'n'
ax2 = axes[1]
sns.lineplot(data=df, x='h', y='wall_time_poly', hue='n', marker='o', ax=ax2, legend=None)
sns.lineplot(data=df, x='h', y='wall_time_ddsim', hue='n', marker='x', ax=ax2, legend=None)
sns.lineplot(data=df, x='h', y='wall_time_aer', hue='n', marker='^', ax=ax2, legend=None)
ax2.set_title('Wall Time vs h')
ax2.set_xlabel('h')
ax2.set_ylabel('Wall Time (s)')
ax2_2 = ax2.twinx()  # Create secondary y-axis
sns.lineplot(data=df, x='h', y='n', ax=ax2_2, color='grey', linestyle='--', marker='D')
ax2_2.set_ylabel('n')

# Improve layout and aesthetics
plt.tight_layout()
plt.show()
````
