```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Graph 1 
n = const, n = 3, h = varying, h = 1 to 29, h_prob = 0.2

```python
# Load data from CSV
data = pd.read_csv("testing.csv")
data.columns = data.columns.str.strip()

# Extract columns
h = data["h"].values
wall_time_poly = data["wall_time_poly"].values
wall_time_ddsim = data["wall_time_ddsim"].values
wall_time_aer = data["wall_time_aer"].values

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(h, wall_time_poly, marker='o', linestyle='-', label='wall_time_poly', color='r')
plt.plot(h, wall_time_ddsim, marker='s', linestyle='-', label='wall_time_ddsim', color='b')
plt.plot(h, wall_time_aer, marker='^', linestyle='-', label='wall_time_aer', color='g')

plt.xlabel("h")
plt.ylabel("Time (seconds)")
plt.title("Wall Time vs h")
plt.yscale("log")  # Log scale for better visualization
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

Extrapolating the wall_time_poly when the value goes to -1

```python
# Load data
df = pd.read_csv("Results/run2/arbitrary_h/program_data_h copy 2.csv")
df.columns = df.columns.str.strip()

for n in [3,30]:
    filtered_df = df[(df["h_prob"] == 0.1) & (df["n"] == n)].reset_index(drop=True)
    h_vals = filtered_df["h"].values

    sim_cols = ["wall_time_poly", "wall_time_ddsim", "wall_time_aer"]
    color_map = {"wall_time_poly": "r", "wall_time_ddsim": "b", "wall_time_aer": "g"}

    plt.figure(figsize=(10, 8))

    for sim in sim_cols:
        y_vals = filtered_df[sim].values
        valid_mask = y_vals != -1

        # Plot original values (with markers)
        sns.lineplot(x=h_vals[valid_mask], y=y_vals[valid_mask],
                     label=sim, marker="o", linestyle="-", color=color_map[sim])

        # Extrapolate from last two valid points if there's any -1
        if not np.all(valid_mask):
            idx = np.where(valid_mask)[0]
            if len(idx) >= 2:
                x1, x2 = h_vals[idx[-2]], h_vals[idx[-1]]
                y1, y2 = y_vals[idx[-2]], y_vals[idx[-1]]
                slope = (y2 - y1) / (x2 - x1)

                # h values where y == -1
                bad_h = h_vals[np.logical_not(valid_mask)]
                extrapolated_y = y2 + slope * (bad_h - x2)

                plt.plot(bad_h, extrapolated_y, linestyle="--", color=color_map[sim])

    plt.yscale("log")
    plt.xlabel("h")
    plt.ylabel("Time (seconds)")
    plt.title(f"Wall Time vs h (n = {n})")
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.tight_layout()
    plt.legend()
    plt.show()
```

```python


# Load data
df = pd.read_csv("Results/run2/arbitrary_h/program_data_h copy 2.csv")
df.columns = df.columns.str.strip()

for h in [10, 19]:
    # Filter data for h and h_prob
    filtered_df = df[(df["h_prob"] == 0.1) & (df["h"] == h)].reset_index(drop=True)
    n_vals = filtered_df["n"].values
    n_vals = n_vals[n_vals <= 45]  # Restrict n to values less than or equal to 45
    filtered_df = filtered_df[filtered_df["n"].isin(n_vals)]  # Filter df to match this range

    sim_cols = ["wall_time_poly", "wall_time_ddsim", "wall_time_aer"]
    # color_map = {"wall_time_poly": "r", "wall_time_ddsim": "b", "wall_time_aer": "g"}

    plt.figure(figsize=(10, 6))

    for sim in sim_cols:
        y_vals = filtered_df[sim].values
        valid_mask = y_vals != -1  # Mask to identify valid values (excluding -1)

        # Plot original values (with markers)
        sns.lineplot(x=n_vals[valid_mask], y=y_vals[valid_mask],
                     label=sim, marker="o", linestyle="-")

        # Extrapolate from last two valid points if there's any -1
        if not np.all(valid_mask):
            idx = np.where(valid_mask)[0]
            if len(idx) >= 2:
                x1, x2 = n_vals[idx[-2]], n_vals[idx[-1]]
                y1, y2 = y_vals[idx[-2]], y_vals[idx[-1]]
                slope = (y2 - y1) / (x2 - x1)

                # n values where y == -1
                bad_n = n_vals[np.logical_not(valid_mask)]
                bad_n = bad_n[bad_n <= 45]  # Ensure we only extrapolate for n <= 45

                # Extrapolated y values
                extrapolated_y = y2 + slope * (bad_n - x2)

                # Plot the extrapolated values (dashed lines)
                plt.plot(bad_n, extrapolated_y, linestyle="--")

    plt.yscale("log")
    plt.xlabel("n")
    plt.ylabel("Time (seconds)")
    plt.title(f"Wall Time vs n (h = {h})")
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.tight_layout()
    plt.legend()
    plt.show()
```
