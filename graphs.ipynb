```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load data from CSV
data = pd.read_csv("testing.csv")
data.columns = data.columns.str.strip()

# Extract columns
h = data["h"].values
wall_time_poly = data["wall_time_poly"].values
wall_time_ddsim = data["wall_time_ddsim"].values
wall_time_aer = data["wall_time_aer"].values

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(h, wall_time_poly, marker='o', linestyle='-', label='wall_time_poly', color='r')
plt.plot(h, wall_time_ddsim, marker='s', linestyle='-', label='wall_time_ddsim', color='b')
plt.plot(h, wall_time_aer, marker='^', linestyle='-', label='wall_time_aer', color='g')

plt.xlabel("h")
plt.ylabel("Time (seconds)")
plt.title("Wall Time vs h")
plt.yscale("log")  # Log scale for better visualization
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
# Load the data
df = pd.read_csv("Results/run2/arbitrary_h/program_data_h.csv")  # Replace with your actual file name
df.columns = df.columns.str.strip()

# Filter rows where h = 15
filtered_df = df[(df["h"] == "15") & (df["h_prob"] == "0.05")]
filtered_df = filtered_df.reset_index(drop=True)

filtered_df = filtered_df.drop(index=range(21))  # Removes rows with index 1 and 3

# Display the result
print(filtered_df.head(30))
```

```python
data = filtered_df.head(30)

# Extract columns and ensure correct data types
n = data["n"].astype(int).values
wall_time_poly = data["wall_time_poly"].astype(float).values
wall_time_ddsim = data["wall_time_ddsim"].astype(float).values
wall_time_aer = data["wall_time_aer"].astype(float).values

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n, wall_time_poly, marker='o', linestyle='-', label='Wall Time Poly', color='r')
plt.plot(n, wall_time_ddsim, marker='s', linestyle='-', label='Wall Time Ddsim', color='b')
plt.plot(n, wall_time_aer, marker='^', linestyle='-', label='Wall Time Aer', color='g')

plt.xlabel("n")
plt.ylabel("Time (seconds)")
plt.title("Wall Time vs n (h=15)")
# plt.yscale("log")  # Log scale for better visualization
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
